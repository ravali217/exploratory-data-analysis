```python

ordinal = []
nominal = []
for col in df.columns:
    if df[col].dtype == 'object':
        nominal.append(col)
    elif df[col].dtype in ['int64', 'float64']:
        ordinal.append(col)
print("\nNominal Attributes:", nominal)
print("Ordinal / Numeric Attributes:", ordinal)

```

```python
ordinal = []
nominal = []
discrete = []
continuous = []

for col in df.columns:

    # NOMINAL (categorical without order)
    if df[col].dtype == 'object':
        nominal.append(col)

    # NUMERIC columns
    elif df[col].dtype in ['int64', 'float64']:

        # DISCRETE → integer values
        if df[col].dtype == 'int64':
            discrete.append(col)

        # CONTINUOUS → decimal values
        if df[col].dtype == 'float64':
            continuous.append(col)

        # ORDINAL → numeric but ordered (you decide manually later)
        ordinal.append(col)

print("Nominal Attributes:", nominal)
print("Ordinal Attributes:", ordinal)
print("Discrete Attributes:", discrete)
print("Continue Attributes:",continuous)

```
